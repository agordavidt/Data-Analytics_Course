## Bar Charts
Bar charts usually present categorical variables, discrete variables or continuous variables grouped in class intervals. They consist of an axis and a series of labelled horizontal or vertical bars. The bars depict frequencies of different values of a variable or simply the different values themselves. 


```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib import style
```

## Vertical Bar Chat
Vertical bar charts are useful to compare different categorical or discrete variables, such as age groups, classes, schools, etc., as long as there are not too many categories to compare. They are also very useful for time series data. The space for labels on the x-axis is small, but ideal for years, minutes, hours or months.

```python
# charts to show numbner of army recruitment from 2010 to date
years = [2010 + x for x in range(14)]

recruitments = [55, 52, 76, 48, 43, 49, 57, 50, 55,57, 67, 55, 49, 50]

# setting the figure size of the plot
plt.figure(figsize=(12, 6))

plt.bar(years, recruitments, color="#abb8c3")
plt.title("Nigerian Army Recruitments, 2010 to 2023")
plt.ylabel("Number of officers")
plt.xlabel("Year")

plt.grid(True)

#choosing a style

plt.show()
```

```python
# to emphasis the highest occuring data value
years = [2010 + x for x in range(14)]

recruitments = [55, 52, 76, 48, 43, 49, 57, 50, 55,57, 67, 55, 49, 50]

style.use("default")

# setting the figure size of the plot
plt.figure(figsize=(12, 6))

bars = plt.bar(years, recruitments, color="#abb8c3")
# show emphasy
bars[2].set_color("#5b1fa2")
plt.title("Nigerian Army Recruitments, 2010 to 2023")
plt.ylabel("Number of officers")
plt.xlabel("Year")

plt.grid(True)

#

plt.show()
```

The grouped bar chart is another effective means of comparing sets of data about the same places or items.

## Horizontal bar charts
The disadvantage of vertical bar charts is that they lack space for text labelling at the foot of each bar. 
When category labels in the chart are too long, the horizontal bar chart is  better for displaying information

```python
# Data for sports practiced by senior secondary school students in Best Academy

years = [2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019]
boys = [110, 185, 240, 285, 305, 310, 315, 315]
girls = [85, 175, 225, 295, 280, 315, 305, 320]


bar_height = 0.35

# Set the bar positions
bar_positions_boys = np.arange(len(years))
bar_positions_girls = bar_positions_boys + bar_height

# Plotting
plt.figure(figsize=(10, 8))



plt.barh(bar_positions_boys, boys, height=0.35, color="#607d8b", label='Boys')
plt.barh(bar_positions_girls, girls, height=0.35, color="#7b1fa2", label='Girls')

# Adding labels and title with custom font size, family, and color
plt.xlabel('Number of Students', color='#3f51b5')
plt.ylabel('Year', color='#3f51b5')
plt.title('Sports Practiced By SSS Students in Abuja, by gender', fontsize=14, fontfamily='serif', color='#3f51b5')

# Adding legend
plt.legend()

# Display the plot
plt.yticks(bar_positions_boys + bar_height/2, years)
#plt.grid(axis='x')
plt.show()
```

## Stacked Bar Charts
The stacked bar chart is a preliminary data analysis tool used to show segments of totals. The stacked bar chart can be very difficult to analyze if too many items are in each stack. 
It can contrast values, but not necessarily in the simplest manner.

```python
names = ['Bello', 'Lola', 'Chima', 'Seaun', 'Ngozi', 'Palang', 'Zichat', 'Hemen']
swimming = [13, 32, 21, 41, 9, 28, 32, 38]
cycling = [50, 53, 28, 14, 81, 47, 40, 24]
running = [37, 15, 51, 45, 10, 25, 28, 38]

# Bar positions
bar_positions = np.arange(len(names))

# Plotting
plt.figure(figsize=(10, 6))

plt.barh(bar_positions, swimming, label='Swimming', color='#263238')
plt.barh(bar_positions, cycling, left=swimming, label='Cycling', color='#607D8B')
plt.barh(bar_positions, running, left=np.array(swimming) + np.array(cycling), label='Running', color='#CFD8DC')

# Adding labels and title
plt.xlabel('Percentage of Time Spent')
plt.ylabel('Names')
plt.title('Percentage of Time Spent in Different Activities')

# Adding legend
plt.legend(loc="upper right")

# Display the plot
plt.yticks(bar_positions, names)
plt.grid(axis='x')
plt.show()
```
