```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import style

```

```python
%matplotlib inline
```

## Box Plot

```python
heights = np.random.normal(172, 8, 200)

plt.boxplot(heights)
plt.show()
```

```python

#style.use("ggplot") #classic #solarize_Light2 #bmh
#style.use("dark_background")
style.use("grayscale")
years = [2014, 2015, 2016, 2017, 2018, 2019, 2020, 2021]

income = [55, 56, 62, 61, 72, 72, 73, 75]


# define the scale
income_ticks = list(range(50,81,2))

plt.plot(years, income)
plt.title("Income of John (in US)")
plt.xlabel("Year")
plt.ylabel("Yearly Income in USD")
plt.yticks(income_ticks, [f"{x}k USD" for x in income_ticks])

plt.show()
```

## Multiple figures in one window

```python
x1, y1 = np.random.random(100), np.random.random(100)
x2, y2 = np.arange(100), np.random.random(100)

plt.figure(1)
plt.scatter(x1, y1)


plt.figure(2)
plt.plot(x2, y2)

plt.show()
```

```python
x = np.arange(100)
style.use("dark_background")


fig, axs = plt.subplots(2, 2)

axs[0,0].plot(x, np.sin(x))
axs[0,0].set_title("sine Wave")
axs[0,0].set_xlabel("Time")
axs[0,0].set_ylabel("Frequency")

axs[0, 1].plot(x, np.cos(x))
axs[0, 1].set_title("Cosine Wave")
axs[0,1].set_xlabel("Time")
axs[0,1].set_ylabel("Frequency")

axs[1, 0].plot(x, np.random.random(100))
axs[1, 0].set_title("Cosine Wave")
axs[1,0].set_xlabel("Time")
axs[1,0].set_ylabel("Frequency")

axs[1, 1].plot(x, np.log(x))
axs[1, 1].set_title("Log Function")
axs[1, 1].set_xlabel("TEST")
axs[1, 1].set_ylabel("Frequency")



# figure can also have a title
fig.suptitle("Four Plots")

# prevent overlapping
plt.tight_layout()

# set parameters
plt.savefig("plot_trans.png", dpi=300, transparent=True, bbox_inches="tight")

plt.show()
```
