```python
# import the matplotlib module
import matplotlib.pyplot as plt

# import numpy as np
import numpy as np
```

```python
# PIE chart
data = np.random.randint(10, 100, 6)
plt.pie(data)
plt.show()
```

```python
genre = ['afrobeat', 'rap', 'gospel', 'fuji', 'highlife']
popularity = [100, 40, 80, 20, 10]
explodes = [0,0,0,0,0.3]
colors = ['#abb8c3','#7b1fa2','#009688', '#b80000', '#fcb900']

plt.pie(popularity, labels=genre, explode=explodes, colors=colors, autopct='%.0f%%')
plt.show()
```

```python
genre = ['afrobeat', 'rap', 'gospel', 'fuji', 'highlife']
popularity = [100, 40, 80, 20, 10]
explodes = [0,0,0,0,0.3]
colors = ['#abb8c3','#7b1fa2','#009688', '#b80000', '#fcb900']

plt.pie(popularity, labels=None, explode=explodes, colors=colors, autopct='%.0f%%')
# Show labels outside the plot the pies // show in legends
plt.legend(labels=genre)
plt.show()
```

```python
# Doughnut Pie plot
circle = plt.Circle( (0,0), 0.7, color='white')

plt.pie(popularity, labels= genre, explode=explodes)
p = plt.gcf()
p.gca().add_artist(circle)


plt.title("Popularity of Highlife Genre Amongs Nigerian Youths")

plt.savefig("genre.png", bbox_inches='tight')
plt.show()
```
