## Basic Styling Functions

```python
# c  = colors
# marker = type of points
# s = size of markers
# alpha = transperency
plt.scatter(x_data, y_data, c='#f00', marker="*", s=150) 
plt.show()
```

```python
# we have more data set / we make the markers transparent
x_data = np.random.random(1000) * 100
y_data = np.random.random(1000) * 100

plt.scatter(x_data, y_data, c='#00f', marker="*", s=200, alpha=0.3) 
plt.show()
```
